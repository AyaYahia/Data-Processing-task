```python
#BY : Aya Yehia Khaled Hamouda, sec 1
outliers=[]
def detect_outliers(dataset):
    
    threshold=3
    mean=np.mean(dataset)
    std=np.std(dataset)
    
    for i in dataset:
        z_score=(i- mean)/std
        if np.abs(z_score) > threshold:
            outliers.append(i)
    return outliers
```

```python
outlier_pt=detect_outliers(dataset)
```
